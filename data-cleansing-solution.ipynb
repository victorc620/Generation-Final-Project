```python
import pandas as pd
import numpy as np
from math import isnan
```

```python
df = pd.read_csv("data-cleansing-exercise/british-library-books.csv")
```

```python
df.head(2)
```

```python
to_drop = [
    'Edition Statement',
    'Corporate Author',
    'Corporate Contributors',
    'Former owner',
    'Engraver',
    'Contributors',
    'Issuance type',
    'Shelfmarks',
    'Flickr URL'
]

df.drop(columns=to_drop, inplace=True)
df.head()
```

```python
df.set_index('Identifier', inplace=True)
df.head()
```

```python
col = 'Date of Publication'

for i in range(len(df[col])):
    current = df[col].iloc[i]

    if "." in str(current):
        df[col].iloc[i] = np.nan
    
    if "?" in str(current):
        df[col].iloc[i] = np.nan
        
    if type(current).__name__ == "float":
        df[col].iloc[i] = np.nan
    
    if str(current).startswith("["):
        df[col].iloc[i] = np.nan

    
```
